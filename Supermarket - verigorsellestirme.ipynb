## Veri Görselleştirme

```python
#Aylara göre alışveriş
plt.figure(figsize=(12,6))
sns.countplot(x=df["month_name"])
plt.title("Aylara göre alışveriş")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df["month_name"],hue=df["Branch"])
```

```python
plt.title("Cinsiyete göre aylık alışveriş sayıları")
sns.countplot(x=df["month_name"],hue=df["Gender"])
```

```python
plt.title("Cinsiyete göre aylık alışveriş sayıları")
sns.countplot(y=df["month_name"],hue=df["Gender"])
plt.ylabel=("months")
plt.show()
```

```python
sns.countplot(x=df["day_name"])
plt.xticks(rotation=45);
```

```python
#En yoğun saatlere
sns.countplot(x=df["hour"])
plt.xticks(rotation=45);
```

```python
df.groupby("hour")["Total"].sum().plot()
```

```python
df["Product line"].value_counts().plot.pie(autopct="&%1.1f%%",explode=[.3,0,0,0,0,0])
```

```python
renkler={"Male":"green","Female":"purple"}
sns.countplot(x=df["Product line"],hue=df["Gender"],palette=renkler)
plt.title("Cinsiyete göre ürün türü alışveriş davranışları")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(16,12))
sns.countplot(x=df["Product line"],hue=df["Branch"])
plt.xticks(rotation=45)
plt.savefig("Sube_urun.png",dpi=300)
```

```python
# Her bir şubenin saatlik ciro
df.groupby(["Branch","hour"])["Total"].sum().unstack(level="hour").plot(kind="bar",figsize=(12,6))
```
